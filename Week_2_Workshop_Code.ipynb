```python
#Interactive Temperature & Clothing Advice
temp = float(input("Enter today's temperature in °C: "))

if temp < 15:
  print("Cold - Wear a jacket")
elif temp < 28:
  print("Warm - T-shirt weather")
else:
  print("Hot - Wear light clothes")
```

```python
#Temperature Check for Several Days
cold_days = 0
warm_days = 0
hot_days = 0

for day in range(1, 4):
  temp = float(input(f"Enter temperature for day {day} in °C: "))

  if temp < 15:
    print("Cold - Wear a jacket")
    cold_days += 1
  elif temp < 28:
    print("Warm - T-shirt weather")
    warm_days += 1
  else:
    print("Hot - Wear light clothes")
    hot_days += 1

  print("\nWeather Summary")
  print("Cold days:", cold_days)
  print("Warm days:", warm_days)
  print("Hot days:", hot_days)
```

```python
#Mini Project - Simple Weather & Event Planner
temp = float(input("Enter today's temperature in C: "))

weather = input("Is it sunny or raining? ").strip().lower()

weekend = input("Is it the weekend? (yes/no): ").strip().lower() == "yes"

if temp < 15:
  category = "Cold"
  advice = "Wear a jacket"
elif temp < 28:
  category = "Warm"
  advice = "T-shirt weather"
else:
  category = "Hot"
  advice = "Wear light clothes"

if weather == "sunny" and weekend:
  activity = "Perfect for a trip to Bondi Beach!"
else:
  activity = "Maybe stay indoors or plan for later."

events = ["Music Festival in Sydney", "Coding Bootcamp Australia", "Farmers Market Brisbane"]

print(f"\nWeather: {category}\nAdvice: {advice}\n{activity}\n")

print("Upcoming Events:")
for i, event in enumerate(events, start=1):
  short_name = event[:10]
  print(f"{i}. {short_name}")

while True:
  choice = input("\nPress 'q' to quit or any other key to see events again: ").lower()
  if choice == "q":
    print("Goodbye!")
    break
  else:
    for i, event in enumerate(events, start=1):
      print(f"{i}. {event[:10]}")
```
